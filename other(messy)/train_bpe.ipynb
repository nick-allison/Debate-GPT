```python
with open("/content/drive/MyDrive/all_debates.txt", "r", encoding='utf-8') as file:
    text = file.read()

text
```

```python
vocab = [x for x in sorted(list(set(text)))]
```

```python
text_list = list(text)

text_list
```

```python
encode_dict = {}
decode_dict = {}

for i in range(len(vocab)):
    encode_dict[vocab[i]] = i
    decode_dict[i] = vocab[i]

next_index = len(vocab)
```

```python
next_index
```

```python
def addToVocab(subw):
    global next_index
    encode_dict[subw] = next_index
    decode_dict[next_index] = subw
    next_index += 1
```

```python
for i in range(len(text_list)):
    text_list[i] = encode_dict[text_list[i]]

text_list
```

```python
byte_pairs = []
byte_pairs = list(zip(text_list, text_list[1:]))

byte_pairs
```

```python
counts_dict = {}

for i in byte_pairs:
      if i not in counts_dict:
          counts_dict[i] = 1
      else:
          counts_dict[i] += 1

counts_dict
```

```python
def updateCorpus(byte_pairs, counts_dict, new_vocab):
    subw = decode_dict[new_vocab[0]] + decode_dict[new_vocab[1]]

    print(subw)

    vocab.append(subw)
    addToVocab(subw)

    new_i = encode_dict[subw]

    for i in byte_pairs:
        for j in range(len(i)-1, -1, -1):
            if i[j] == new_vocab:
                if j != 0:
                    if counts_dict[i[j-1]] <= 1:
                        del counts_dict[i[j-1]]
                    else:
                        counts_dict[i[j-1]] -= 1

                    i[j-1] = (i[j-1][0],new_i)

                    if i[j-1] not in counts_dict:
                        counts_dict[i[j-1]] = 1
                    else:
                        counts_dict[i[j-1]] += 1
                if j != len(i) - 1:
                    if counts_dict[i[j+1]] <= 1:
                        del counts_dict[i[j+1]]
                    else:
                        counts_dict[i[j+1]] -= 1

                    i[j+1] = (new_i, i[j+1][1])

                    if i[j+1] not in counts_dict:
                        counts_dict[i[j+1]] = 1
                    else:
                        counts_dict[i[j+1]] += 1

                i.pop(j)

    del counts_dict[new_vocab]
```

```python
K = 10000
VOCAB_OUTPUT_FILE = '/content/drive/MyDrive/vocab.txt'
RESULT_OUTPUT_FILE = '/content/drive/MyDrive/result.txt'
```

```python
for _ in range(K):
    if len(counts_dict) != 0:
        updateCorpus(byte_pairs, counts_dict, max(counts_dict, key=counts_dict.get))
    else:
        print("No further sub-word tokenization possible")
```

```python
def replace_substring(text, mapping):
    """
    Replaces each occurrence of a substring in `mapping` with its corresponding value.

    :param text: The input string
    :param mapping: Dictionary of substrings to replace with numbers
    :return: List where matches are replaced and other parts are kept as strings
    """
    print("func")
    # Create regex pattern to match all keys in mapping
    pattern = "|".join(re.escape(k) for k in list(mapping.keys())[::-1])

    # Split string but keep the delimiters
    parts = re.split(f'({pattern})', text)

    # Replace found substrings with their corresponding numbers
    result = [mapping[p] if p in mapping else p for p in parts if p]

    return result
```
